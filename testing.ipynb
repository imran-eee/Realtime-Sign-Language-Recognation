```python
import cv2
from cvzone.HandTrackingModule import HandDetector
from tensorflow.keras.models import load_model
import numpy as np
import math

cap = cv2.VideoCapture(0)
detector = HandDetector(maxHands=1)
model = load_model("model-016.h5")

offset = 20
imgSize = 224
counter = 0

labels = ['call me','Hello', 'no','Ok','thank you','Victory','yes']

while True:
    success, img = cap.read()
    imgOutput = img.copy()
    hands, img = detector.findHands(img)
    if hands:
        hand = hands[0]
        x, y, w, h = hand['bbox']
 
        imgCrop = img[y - offset:y + h + offset, x - offset:x + w + offset]
        if imgCrop.size != 0:
            imgResize = cv2.resize(imgCrop, (imgSize, imgSize))
            imgGray = cv2.cvtColor(imgResize, cv2.COLOR_BGR2GRAY)
            imgInput = np.reshape(imgGray, (imgSize, imgSize, 1)) / 255.0

            prediction = model.predict(np.array([imgInput]))
            prediction = np.argmax(prediction)
            gesture = labels[prediction]
            
            cv2.putText(imgOutput, gesture, (x+20,y-20), cv2.FONT_HERSHEY_PLAIN, 3, (255,0,255), 3)

            cv2.rectangle(imgOutput, (x - offset, y - offset),
                          (x + w + offset, y + h + offset), (255, 0, 255), 2)
 
    cv2.imshow("Image", imgOutput)
    cv2.waitKey(1)
```
